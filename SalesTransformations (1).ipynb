```python
# Load sales data from bronze layer
df = spark.read.format("parquet").load("Files/bronze/sales.parquet")
display(df)



```

```python
# Add a new column: total_sales = quantity * price
from pyspark.sql.functions import col

df_transformed = df.withColumn("total_sales", col("quantity") * col("price"))


df_transformed = df_transformed.dropDuplicates()
```

```python
#Write to Silver as Parquet
df_transformed.write.mode("overwrite").format("parquet").save("Files/silver/sales_transformed")
```

```python
#Load back to verify
df_silver = spark.read.format("parquet").load("Files/silver/sales_transformed")
display(df_silver)
```

```python
from pyspark.sql.functions import col

df_gold = df.withColumn("sales_amount", col("quantity") * col("price"))
```

```python
df_gold.write.mode("overwrite").format("parquet").save("Files/gold/sales_enriched")
```

```python
df_check = spark.read.format("parquet").load("Files/gold/sales_enriched")
display(df_check)
```
